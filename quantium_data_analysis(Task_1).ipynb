```python
#removing the outlier from transaction dataset
transaction_data = transaction_data[transaction_data['LYLTY_CARD_NBR'] != 226000]
```

```python
transaction_data["DATE"] = pd.to_datetime(
    transaction_data["DATE"],
    origin="1899-12-30",
    unit="D"
)
```

```python
transaction_data.head()
```

```python
#checking for the Maximum and Minimum date time in transaction dataset
print("The maximum date is : ",transaction_data['DATE'].max())
print("The minimum date is : ",transaction_data['DATE'].min())
```

```python
# extracting Brand Name (first word)
transaction_data["BRAND_NAME"] = transaction_data["PROD_NAME"].str.extract(r'^([A-Za-z0-9&\'\-]+)')
transaction_data
```

```python
# Pack Size
pack_pattern = r'(\d+(?:\.\d+)?\s?(?:kg|g|mg|l|ml|oz|lb|cl|pk|pcs|pc|ct))\b'

transaction_data["PACK_SIZE"] = (
    transaction_data["PROD_NAME"]
      .str.extract(pack_pattern, flags=re.IGNORECASE)[0]
      .str.replace(r"\s+", "", regex=True)
      .str.lower()
)
```

```python
#creating a mapping
brand_mapping = {
    'Infzns' : 'Infuzions',
    'Dorito' : 'Doritos',
    'Smith' : 'Smiths',
    'Grain' : 'GrnWves',
    'Red' : 'RRD',
    'WW' : 'Woolworths',
    'Snbts' : 'Sunbites',
    'Natural' : 'Natural Chip Co',
    'Old' : 'Old El Paso'
}
```

```python
#standarizing the names
transaction_data["BRAND_NAME"] = transaction_data["BRAND_NAME"].replace(brand_mapping)
```

```python
transaction_data
```

```python
transaction_data.to_excel("Final_transaction_data.xlsx",index=False)
```

```python
#Merging Final transaction datatset with purchase behaviour dataset
merged_data = pd.merge(
    transaction_data,
    purchase_behaviour_data,
    on="LYLTY_CARD_NBR",
    how="left"
)
```

```python
merged_data.shape
```

```python
merged_data.isnull().sum()
```

```python
merged_data.head()
```

```python
#adding a metrics (Price Per Unit)
merged_data["PRICE_PER_UNIT"] = (
    merged_data["TOT_SALES"] / merged_data["PROD_QTY"]
)
```

```python
merged_data
```

```python
#which lifestage has generated more revenue
merged_data.groupby("LIFESTAGE")["TOT_SALES"].sum().sort_values(ascending=False)
```

```python
#which customer category has generated more sales
merged_data.groupby("PREMIUM_CUSTOMER")["TOT_SALES"].sum()
```

```python
#who has generated more sales in both category
merged_data.groupby(
    ["LIFESTAGE","PREMIUM_CUSTOMER"]
)["TOT_SALES"].sum()
```

```python
# unique customers in each segments
merged_data.groupby(
    ["LIFESTAGE","PREMIUM_CUSTOMER"]
)["LYLTY_CARD_NBR"].nunique()
```

```python
#average spend per customers
merged_data.groupby(
    ['LIFESTAGE','PREMIUM_CUSTOMER']
)['TOT_SALES'].mean()
```

```python
#Average quantity purchased
merged_data.groupby(
    ['LIFESTAGE','PREMIUM_CUSTOMER']
)['PROD_QTY'].mean()
```

```python
#which brand is more popular
merged_data.groupby('BRAND_NAME')['TOT_SALES'].sum().sort_values(ascending=False)
```

```python
#pack size popular
merged_data.groupby('PACK_SIZE')['TOT_SALES'].sum().sort_values(ascending=False)
```
